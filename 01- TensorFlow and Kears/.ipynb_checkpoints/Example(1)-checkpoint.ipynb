```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist
```

```python
(x_train , y_train) , (x_test , y_test) = mnist.load_data()
x_train , x_test = x_train / 255.0 , x_test /255.0
```

```python
print(f"The X_Train_shape :: {x_train.shape}\nFist_sample :: {x_train[5]}\n************************************")
print(f"The Y_Train_shape :: {y_train.shape}\nFist_sample :: {y_train[5]}\n****************************************")
print(f"The X_Test_shape :: {x_test.shape}\nFist_sample :: {x_test[5]}\n********************************************")
print(f"The Y_Test_shape :: {y_test.shape}\nFist_sample :: {y_test[5]}\n***********************************************")
```

```python
# Bulid NN
Model = tf.keras.models.Sequential()
```

```python
# add Input Features as flatten
Model.add(tf.keras.layers.Flatten())
```

```python
# # add hidden layers and use relue as activation func
Model.add(tf.keras.layers.Dense(128 , activation = tf.nn.relu))
# activation = (relu , linear elu , sigmoid , tanch , softmax) Bust with hideen layer most efficient relu
```

```python
# add dropbox to reduce num of connection and overfitting risk
Model.add(tf.keras.layers.Dropout(0.2))
```

```python
# add output layers its nerron equalize number of classes
Model.add(tf.keras.layers.Dense(10 , activation = tf.nn.softmax))
```

```python
Model.compile(optimizer="adam" , loss="sparse_categorical_crossentropy" , metrics=["accuracy"])
# Loss if reg = "mse" if Binary_classes = "sparse_binary_crossentropy" if Multi "sparse_categorical_crossentropy" 
```

```python
Model.fit(x_train , y_train , epochs=7)
```

```python
out_put = Model.predict(x_test)
for i in range(10) :
    print(out_put[i])
```

```python
val_loss , val_acc = Model.evaluate(x_test , y_test)
```

```python
Model.save("model.model")
```
