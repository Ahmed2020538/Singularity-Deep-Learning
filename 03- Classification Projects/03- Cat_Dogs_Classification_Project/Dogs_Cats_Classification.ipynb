```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
import tensorflow as tf
from tensorflow import keras
```

```python
Data_dir = "PetImages"
Categories = ["Dog" , "Cat"]
for category in Categories :
    path = os.path.join(Data_dir , category)
    x = 0
    for img in os.listdir(path) :
        x+=1
        img_arr = cv2.imread(os.path.join(path , img) , cv2.IMREAD_IGNORE_ORIENTATION)
        plt.imshow(img_arr , cmap="gray")
        plt.show()
        if x == 20 :
            break
```

```python
img_arr.shape
```

```python
img_size = 90
img_Arr = cv2.resize(img_arr , (img_size , img_size))
plt.imshow(img_Arr , cmap="gray")
plt.show()
```

```python
Training_data = []
def create_Traing_data() :
    for category in Categories :
        path = os.path.join(Data_dir , category)
        classes_label = Categories.index(category)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_GRAYSCALE)  # convert to array
                new_arr = cv2.resize(img_arr, (img_size, img_size))  # resize to normalize data size
                Training_data.append([new_arr, classes_label])  # add this to our training_data
            except Exception as e:  # in the interest in keeping the output clean...
                pass

create_Traing_data()
```

```python
len(Training_data)
```

```python
import random
random.shuffle(Training_data)
for sample in Training_data[:10]:
    print(sample[1])
```

```python
X = []
y = []
img_size =90
for features,label in Training_data:
    X.append(features)
    y.append(label)

print(X[0].reshape(-1, img_size, img_size, 1))

X = np.array(X).reshape(-1, img_size, img_size)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(img_size,img_size)),
    keras.layers.Dense(256, activation=tf.nn.relu),
    #keras.layers.Dropout(0.2) ,
    keras.layers.Dense(256, activation=tf.nn.relu),
    keras.layers.Dense(2, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10)
```
