```python
# drop useless features
df.drop("Unnamed: 0" , axis=1 , inplace=True)
```

```python
# identify output values
df['Species'].unique()
```

```python
# Handling Text data 
df = pd.get_dummies(df, columns=['Species'])
df
```

```python
# split data into inpute features and output features
x = df[["Sepal.Length" , "Sepal.Width" , "Petal.Length" , "Petal.Width"]]
y = df[["Species_setosa" , "Species_versicolor" ,"Species_virginica" ]]
```

```python
# Split data into train , test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
# Scalling data in normal distribution
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x.shape
```

```python
y.shape[1]
```

* Build Model Structure

```python
model = keras.models.Sequential([
    keras.layers.Dense(50 , input_shape = [x.shape[1]], activation = "relu"),
    keras.layers.Dropout(0.2) , 
    keras.layers.Dense(y.shape[1] , activation = "softmax")
])
```

```python
model.summary()
```

```python
model.compile(optimizer= "adam", loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping

lrd = ReduceLROnPlateau(monitor = 'val_loss',
                         patience = 10,
                         verbose = 1,
                         factor = 0.75,
                         min_lr = 1e-10)

mcp = ModelCheckpoint('model.h5')

es = EarlyStopping(verbose=1, patience=100)
```

```python
# Fitting Model
history = model.fit(x=x_train, y=y_train, epochs=500, callbacks=[lrd, mcp, es], batch_size=32, validation_split=0.1)
```

```python
# Predict new values
y_pred = model.predict(x_test)
y_pred
```

```python
# Predict y classes
y_pred = model.predict_classes(x_test)
y_pred
```

```python
# Evaluate model performance ::
model.evaluate(x_test, y_test)
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
